```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import load_model

warnings.filterwarnings("ignore")

# =========================
# PATH CONFIGURATION
# =========================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Juvenile Crime Growth Prediction using Bio-Inspired Optimization"
CSV_PATH = os.path.join(BASE_DIR, "Table_30_1976.csv")

os.makedirs(BASE_DIR, exist_ok=True)

PREFIX = "bio_"

# =========================
# LOAD DATA
# =========================

df = pd.read_csv(CSV_PATH)

df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print(df.head())

# =========================
# DATA CLEANING
# =========================

df = df.dropna()

target_column = "Total juvenile crime cases under IPC"

if target_column not in df.columns:
    raise ValueError(f"Target column not found: {target_column}")

# Juvenile Crime Growth %
df["Juvenile Crime Growth Rate"] = df[target_column].pct_change() * 100
df["Juvenile Crime Growth Rate"] = df["Juvenile Crime Growth Rate"].fillna(0)

# Final target
target_column = "Juvenile Crime Growth Rate"

# Remove unnecessary serial column
if "SL. No." in df.columns:
    df = df.drop(columns=["SL. No."])

# =========================
# FEATURE TARGET SPLIT
# =========================

X = df.drop(columns=[target_column])
y = df[target_column]

X = X.select_dtypes(include=[np.number])

feature_names = X.columns.tolist()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# =========================
# FITNESS FUNCTION
# =========================

def fitness_function(mask):
    mask = np.array(mask).astype(bool)

    if mask.sum() == 0:
        return 999999

    X_selected = X_scaled[:, mask]

    X_train, X_test, y_train, y_test = train_test_split(
        X_selected, y, test_size=0.30, random_state=42
    )

    model = RandomForestRegressor(
        n_estimators=100,
        random_state=42
    )

    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    mse = mean_squared_error(y_test, preds)

    return mse


# =========================
# PSO FEATURE SELECTION
# =========================

def pso_feature_selection(num_particles=20, max_iter=30):
    num_features = X_scaled.shape[1]

    particles = np.random.randint(0, 2, (num_particles, num_features))
    velocity = np.random.rand(num_particles, num_features)

    personal_best = particles.copy()
    personal_best_scores = np.array([fitness_function(p) for p in particles])

    global_best = personal_best[np.argmin(personal_best_scores)]
    global_best_score = np.min(personal_best_scores)

    fitness_curve = []

    for iteration in range(max_iter):
        for i in range(num_particles):
            r1 = np.random.rand(num_features)
            r2 = np.random.rand(num_features)

            velocity[i] = (
                0.7 * velocity[i]
                + 1.5 * r1 * (personal_best[i] - particles[i])
                + 1.5 * r2 * (global_best - particles[i])
            )

            sigmoid = 1 / (1 + np.exp(-velocity[i]))
            particles[i] = np.where(np.random.rand(num_features) < sigmoid, 1, 0)

            score = fitness_function(particles[i])

            if score < personal_best_scores[i]:
                personal_best[i] = particles[i]
                personal_best_scores[i] = score

        global_best = personal_best[np.argmin(personal_best_scores)]
        global_best_score = np.min(personal_best_scores)

        fitness_curve.append(global_best_score)

        print(f"Iteration {iteration + 1}/{max_iter}, Best MSE: {global_best_score}")

    return global_best, fitness_curve


best_mask, fitness_curve = pso_feature_selection()

selected_features = [
    feature for feature, selected in zip(feature_names, best_mask) if selected == 1
]

print("Selected Features:", selected_features)

# =========================
# TRAIN FINAL MODEL
# =========================

X_selected = X_scaled[:, best_mask.astype(bool)]

X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y, test_size=0.30, random_state=42
)

rf_model = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

rf_model.fit(X_train, y_train)

rf_predictions = rf_model.predict(X_test)

mae = mean_absolute_error(y_test, rf_predictions)
mse = mean_squared_error(y_test, rf_predictions)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, rf_predictions)

# =========================
# DEEP LEARNING MODEL H5
# =========================

dl_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_selected.shape[1],)),
    Dense(32, activation="relu"),
    Dense(16, activation="relu"),
    Dense(1)
])

dl_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

history = dl_model.fit(
    X_train,
    y_train,
    epochs=100,
    batch_size=2,
    validation_data=(X_test, y_test),
    verbose=0
)

dl_predictions = dl_model.predict(X_test).flatten()

dl_mae = mean_absolute_error(y_test, dl_predictions)
dl_mse = mean_squared_error(y_test, dl_predictions)
dl_rmse = np.sqrt(dl_mse)
dl_r2 = r2_score(y_test, dl_predictions)

# =========================
# SAVE MODELS
# =========================

joblib.dump(rf_model, os.path.join(BASE_DIR, PREFIX + "random_forest_model.pkl"))
joblib.dump(scaler, os.path.join(BASE_DIR, PREFIX + "scaler.pkl"))

dl_model.save(os.path.join(BASE_DIR, PREFIX + "deep_learning_model.h5"))

# =========================
# SAVE CSV FILES
# =========================

selected_feature_df = pd.DataFrame({
    "Selected Features": selected_features
})

selected_feature_df.to_csv(
    os.path.join(BASE_DIR, PREFIX + "selected_features.csv"),
    index=False
)

prediction_df = pd.DataFrame({
    "Actual Growth Rate": y_test.values,
    "RF Predicted Growth Rate": rf_predictions,
    "DL Predicted Growth Rate": dl_predictions
})

prediction_df.to_csv(
    os.path.join(BASE_DIR, PREFIX + "predictions.csv"),
    index=False
)

result_df = pd.DataFrame({
    "Model": ["Random Forest", "Deep Learning"],
    "MAE": [mae, dl_mae],
    "MSE": [mse, dl_mse],
    "RMSE": [rmse, dl_rmse],
    "R2 Score": [r2, dl_r2]
})

result_df.to_csv(
    os.path.join(BASE_DIR, PREFIX + "model_results.csv"),
    index=False
)

# =========================
# SAVE JSON SUMMARY
# =========================

summary = {
    "project_name": "Juvenile Crime Growth Prediction using Bio-Inspired Optimization",
    "author": "Sagnik Patra",
    "dataset": CSV_PATH,
    "optimization_algorithm": "Particle Swarm Optimization",
    "machine_learning_model": "Random Forest Regressor",
    "deep_learning_model": "Dense Neural Network",
    "target_column": target_column,
    "selected_features": selected_features,
    "random_forest_metrics": {
        "MAE": float(mae),
        "MSE": float(mse),
        "RMSE": float(rmse),
        "R2 Score": float(r2)
    },
    "deep_learning_metrics": {
        "MAE": float(dl_mae),
        "MSE": float(dl_mse),
        "RMSE": float(dl_rmse),
        "R2 Score": float(dl_r2)
    }
}

with open(os.path.join(BASE_DIR, PREFIX + "result_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

# =========================
# SAVE YAML CONFIG
# =========================

config = {
    "project": "Juvenile Crime Growth Prediction using Bio-Inspired Optimization",
    "author": "Sagnik Patra",
    "base_directory": BASE_DIR,
    "dataset_path": CSV_PATH,
    "target_column": target_column,
    "optimization": {
        "algorithm": "PSO",
        "particles": 20,
        "iterations": 30
    },
    "models": {
        "machine_learning": "RandomForestRegressor",
        "deep_learning": "Dense Neural Network"
    },
    "outputs": {
        "h5_model": PREFIX + "deep_learning_model.h5",
        "pkl_model": PREFIX + "random_forest_model.pkl",
        "scaler": PREFIX + "scaler.pkl",
        "json_summary": PREFIX + "result_summary.json",
        "yaml_config": PREFIX + "config.yaml",
        "predictions": PREFIX + "predictions.csv",
        "results": PREFIX + "model_results.csv"
    }
}

with open(os.path.join(BASE_DIR, PREFIX + "config.yaml"), "w") as f:
    yaml.dump(config, f, sort_keys=False)

# =========================
# SAVE GRAPHS
# =========================

plt.figure(figsize=(10, 6))
plt.plot(fitness_curve, marker="o")
plt.title("PSO Optimization Fitness Curve")
plt.xlabel("Iteration")
plt.ylabel("Best MSE")
plt.grid(True)
plt.savefig(os.path.join(BASE_DIR, PREFIX + "optimization_fitness_graph.png"))
plt.close()

plt.figure(figsize=(10, 6))
plt.plot(y_test.values, marker="o", label="Actual")
plt.plot(rf_predictions, marker="x", label="RF Predicted")
plt.title("Actual vs Predicted Juvenile Crime Growth Rate")
plt.xlabel("Test Sample")
plt.ylabel("Growth Rate")
plt.legend()
plt.grid(True)
plt.savefig(os.path.join(BASE_DIR, PREFIX + "rf_actual_vs_predicted_graph.png"))
plt.close()

plt.figure(figsize=(10, 6))
plt.bar(result_df["Model"], result_df["R2 Score"])
plt.title("Model R2 Score Comparison")
plt.xlabel("Model")
plt.ylabel("R2 Score")
plt.grid(True)
plt.savefig(os.path.join(BASE_DIR, PREFIX + "model_comparison_graph.png"))
plt.close()

plt.figure(figsize=(10, 6))
plt.bar(selected_features, np.ones(len(selected_features)))
plt.title("Selected Features using PSO")
plt.xlabel("Features")
plt.ylabel("Selected")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, PREFIX + "selected_features_graph.png"))
plt.close()

# =========================
# FINAL MESSAGE
# =========================

print("\nAll files generated successfully!")
print("Saved in:", BASE_DIR)

print("\nGenerated Files:")
print(PREFIX + "random_forest_model.pkl")
print(PREFIX + "scaler.pkl")
print(PREFIX + "deep_learning_model.h5")
print(PREFIX + "config.yaml")
print(PREFIX + "result_summary.json")
print(PREFIX + "predictions.csv")
print(PREFIX + "model_results.csv")
print(PREFIX + "selected_features.csv")
print(PREFIX + "optimization_fitness_graph.png")
print(PREFIX + "rf_actual_vs_predicted_graph.png")
print(PREFIX + "model_comparison_graph.png")
print(PREFIX + "selected_features_graph.png")
```
