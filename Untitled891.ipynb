```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    accuracy_score,
    classification_report,
    confusion_matrix
)

warnings.filterwarnings("ignore")

# ==========================================================
# PATH SETUP
# ==========================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Juvenile Crime Growth Prediction using Bio-Inspired Optimization"
CSV_PATH = os.path.join(BASE_DIR, "Table_30_1976.csv")

PREFIX = "juvenile_"

os.makedirs(BASE_DIR, exist_ok=True)

# ==========================================================
# LOAD DATA
# ==========================================================

df = pd.read_csv(CSV_PATH)
df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print(df.head())
print("\nColumns:")
print(df.columns)

# ==========================================================
# CLEAN DATA
# ==========================================================

for col in df.columns:
    if df[col].dtype == "object":
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(",", "", regex=False)
            .str.replace("%", "", regex=False)
        )
        df[col] = pd.to_numeric(df[col], errors="ignore")

df = df.dropna()

# ==========================================================
# TARGET COLUMN
# ==========================================================

juvenile_col = "Total juvenile crime cases under IPC"

if juvenile_col not in df.columns:
    raise ValueError(f"Column not found: {juvenile_col}")

df["Juvenile Crime Growth Rate"] = df[juvenile_col].pct_change() * 100
df["Juvenile Crime Growth Rate"] = df["Juvenile Crime Growth Rate"].fillna(0)

# ==========================================================
# CREATE CLASSES
# ==========================================================

def classify_growth(value):
    if value < 0:
        return "Decline"
    elif value <= 5:
        return "Low Growth"
    elif value <= 15:
        return "Medium Growth"
    else:
        return "High Growth"

df["Growth Class"] = df["Juvenile Crime Growth Rate"].apply(classify_growth)

# ==========================================================
# SAVE CLASSES CSV
# ==========================================================

classes_csv = os.path.join(BASE_DIR, PREFIX + "classes.csv")

df[["Year", "Juvenile Crime Growth Rate", "Growth Class"]].to_csv(
    classes_csv,
    index=False
)

# ==========================================================
# FEATURE AND TARGET SETUP
# ==========================================================

X = df.drop(columns=["Juvenile Crime Growth Rate", "Growth Class"])
X = X.select_dtypes(include=[np.number])

y_reg = df["Juvenile Crime Growth Rate"]
y_cls = df["Growth Class"]

feature_names = X.columns.tolist()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ==========================================================
# SIMPLE PSO FEATURE SELECTION
# ==========================================================

def fitness_function(mask):
    mask = np.array(mask).astype(bool)

    if mask.sum() == 0:
        return 999999

    X_selected = X_scaled[:, mask]

    X_train, X_test, y_train, y_test = train_test_split(
        X_selected,
        y_reg,
        test_size=0.3,
        random_state=42
    )

    model = RandomForestRegressor(
        n_estimators=100,
        random_state=42
    )

    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    return mean_squared_error(y_test, pred)


def pso_feature_selection(num_particles=20, max_iter=30):
    num_features = X_scaled.shape[1]

    particles = np.random.randint(0, 2, (num_particles, num_features))
    velocity = np.random.rand(num_particles, num_features)

    personal_best = particles.copy()
    personal_best_scores = np.array([fitness_function(p) for p in particles])

    global_best = personal_best[np.argmin(personal_best_scores)]
    global_best_score = np.min(personal_best_scores)

    fitness_curve = []

    for iteration in range(max_iter):
        for i in range(num_particles):
            r1 = np.random.rand(num_features)
            r2 = np.random.rand(num_features)

            velocity[i] = (
                0.7 * velocity[i]
                + 1.5 * r1 * (personal_best[i] - particles[i])
                + 1.5 * r2 * (global_best - particles[i])
            )

            sigmoid = 1 / (1 + np.exp(-velocity[i]))

            particles[i] = np.where(
                np.random.rand(num_features) < sigmoid,
                1,
                0
            )

            score = fitness_function(particles[i])

            if score < personal_best_scores[i]:
                personal_best[i] = particles[i]
                personal_best_scores[i] = score

        global_best = personal_best[np.argmin(personal_best_scores)]
        global_best_score = np.min(personal_best_scores)

        fitness_curve.append(global_best_score)

        print(f"Iteration {iteration + 1}/{max_iter} | Best MSE: {global_best_score}")

    return global_best, fitness_curve


best_mask, fitness_curve = pso_feature_selection()

if best_mask.sum() == 0:
    best_mask = np.ones(len(feature_names))

selected_features = [
    feature for feature, selected in zip(feature_names, best_mask)
    if selected == 1
]

X_selected = X_scaled[:, best_mask.astype(bool)]

# ==========================================================
# SAVE SELECTED FEATURES
# ==========================================================

selected_features_df = pd.DataFrame({
    "Selected Features": selected_features
})

selected_features_df.to_csv(
    os.path.join(BASE_DIR, PREFIX + "selected_features.csv"),
    index=False
)

# ==========================================================
# REGRESSION MODEL
# ==========================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_selected,
    y_reg,
    test_size=0.3,
    random_state=42
)

reg_model = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

reg_model.fit(X_train, y_train)

reg_pred = reg_model.predict(X_test)

mae = mean_absolute_error(y_test, reg_pred)
mse = mean_squared_error(y_test, reg_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, reg_pred)

reg_accuracy = max(0, r2) * 100

# ==========================================================
# CLASSIFICATION MODEL
# ==========================================================

label_encoder = LabelEncoder()
y_cls_encoded = label_encoder.fit_transform(y_cls)

try:
    Xc_train, Xc_test, yc_train, yc_test = train_test_split(
        X_selected,
        y_cls_encoded,
        test_size=0.3,
        random_state=42,
        stratify=y_cls_encoded
    )
except:
    Xc_train, Xc_test, yc_train, yc_test = train_test_split(
        X_selected,
        y_cls_encoded,
        test_size=0.3,
        random_state=42
    )

cls_model = RandomForestClassifier(
    n_estimators=300,
    random_state=42
)

cls_model.fit(Xc_train, yc_train)

cls_pred = cls_model.predict(Xc_test)

class_accuracy = accuracy_score(yc_test, cls_pred) * 100

classification_report_dict = classification_report(
    yc_test,
    cls_pred,
    target_names=label_encoder.classes_,
    output_dict=True,
    zero_division=0
)

cm = confusion_matrix(yc_test, cls_pred)

# ==========================================================
# SAVE MODELS
# ==========================================================

joblib.dump(reg_model, os.path.join(BASE_DIR, PREFIX + "regression_model.pkl"))
joblib.dump(cls_model, os.path.join(BASE_DIR, PREFIX + "classification_model.pkl"))
joblib.dump(scaler, os.path.join(BASE_DIR, PREFIX + "scaler.pkl"))
joblib.dump(label_encoder, os.path.join(BASE_DIR, PREFIX + "label_encoder.pkl"))

# ==========================================================
# SAVE PREDICTION CSV
# ==========================================================

prediction_df = pd.DataFrame({
    "Actual Growth Rate": y_test.values,
    "Predicted Growth Rate": reg_pred
})

prediction_csv = os.path.join(BASE_DIR, PREFIX + "prediction.csv")

prediction_df.to_csv(
    prediction_csv,
    index=False
)

# ==========================================================
# SAVE CLASS PREDICTION CSV
# ==========================================================

class_prediction_df = pd.DataFrame({
    "Actual Class": label_encoder.inverse_transform(yc_test),
    "Predicted Class": label_encoder.inverse_transform(cls_pred)
})

class_prediction_df.to_csv(
    os.path.join(BASE_DIR, PREFIX + "class_prediction.csv"),
    index=False
)

# ==========================================================
# SAVE RESULT CSV
# ==========================================================

result_df = pd.DataFrame({
    "Model": [
        "Random Forest Regressor",
        "Random Forest Classifier"
    ],
    "MAE": [
        mae,
        np.nan
    ],
    "MSE": [
        mse,
        np.nan
    ],
    "RMSE": [
        rmse,
        np.nan
    ],
    "R2 Score": [
        r2,
        np.nan
    ],
    "Accuracy": [
        reg_accuracy,
        class_accuracy
    ]
})

result_csv = os.path.join(BASE_DIR, PREFIX + "result.csv")

result_df.to_csv(
    result_csv,
    index=False
)

# ==========================================================
# SAVE CLASSIFICATION REPORT CSV
# ==========================================================

classification_report_df = pd.DataFrame(classification_report_dict).transpose()

classification_report_df.to_csv(
    os.path.join(BASE_DIR, PREFIX + "classification_report.csv")
)

# ==========================================================
# SAVE JSON SUMMARY
# ==========================================================

summary = {
    "project_name": "Juvenile Crime Growth Prediction using Bio-Inspired Optimization",
    "author": "Sagnik Patra",
    "dataset_path": CSV_PATH,
    "selected_features": selected_features,
    "regression_metrics": {
        "MAE": float(mae),
        "MSE": float(mse),
        "RMSE": float(rmse),
        "R2 Score": float(r2),
        "Accuracy": float(reg_accuracy)
    },
    "classification_metrics": {
        "Accuracy": float(class_accuracy),
        "Classes": label_encoder.classes_.tolist()
    }
}

with open(os.path.join(BASE_DIR, PREFIX + "summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

# ==========================================================
# SAVE YAML CONFIG
# ==========================================================

config = {
    "project": "Juvenile Crime Growth Prediction using Bio-Inspired Optimization",
    "author": "Sagnik Patra",
    "dataset": CSV_PATH,
    "output_directory": BASE_DIR,
    "prefix": PREFIX,
    "target_regression": "Juvenile Crime Growth Rate",
    "target_classification": "Growth Class",
    "optimization_algorithm": "PSO",
    "models": {
        "regression": "RandomForestRegressor",
        "classification": "RandomForestClassifier"
    }
}

with open(os.path.join(BASE_DIR, PREFIX + "config.yaml"), "w") as f:
    yaml.dump(config, f, sort_keys=False)

# ==========================================================
# FUNCTION TO SAVE AND SHOW GRAPH
# ==========================================================

def save_and_show(filename):
    plt.tight_layout()
    plt.savefig(os.path.join(BASE_DIR, filename), dpi=300)
    plt.show()
    plt.close()

# ==========================================================
# 1. ACCURACY GRAPH
# ==========================================================

plt.figure(figsize=(9, 6))
plt.bar(result_df["Model"], result_df["Accuracy"])
plt.title("Accuracy Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy (%)")
plt.xticks(rotation=20)
save_and_show(PREFIX + "accuracy_graph.png")

# ==========================================================
# 2. HEATMAP
# ==========================================================

corr = df.select_dtypes(include=[np.number]).corr()

plt.figure(figsize=(12, 8))
plt.imshow(corr, aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
save_and_show(PREFIX + "heatmap.png")

# ==========================================================
# 3. COMPARISON GRAPH
# ==========================================================

plt.figure(figsize=(10, 6))
plt.plot(y_test.values, marker="o", label="Actual Growth Rate")
plt.plot(reg_pred, marker="x", label="Predicted Growth Rate")
plt.title("Actual vs Predicted Comparison Graph")
plt.xlabel("Test Sample")
plt.ylabel("Growth Rate")
plt.legend()
plt.grid(True)
save_and_show(PREFIX + "comparison_graph.png")

# ==========================================================
# 4. RESULT GRAPH
# ==========================================================

plt.figure(figsize=(10, 6))
plt.bar(["MAE", "MSE", "RMSE", "R2 Score"], [mae, mse, rmse, r2])
plt.title("Regression Result Graph")
plt.xlabel("Metric")
plt.ylabel("Value")
save_and_show(PREFIX + "result_graph.png")

# ==========================================================
# 5. PREDICTION GRAPH
# ==========================================================

plt.figure(figsize=(10, 6))
plt.plot(prediction_df["Actual Growth Rate"], marker="o", label="Actual")
plt.plot(prediction_df["Predicted Growth Rate"], marker="x", label="Predicted")
plt.title("Prediction Graph")
plt.xlabel("Sample")
plt.ylabel("Juvenile Crime Growth Rate")
plt.legend()
plt.grid(True)
save_and_show(PREFIX + "prediction_graph.png")

# ==========================================================
# 6. CLASS GRAPH
# ==========================================================

class_counts = df["Growth Class"].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(class_counts.index, class_counts.values)
plt.title("Growth Class Graph")
plt.xlabel("Class")
plt.ylabel("Count")
save_and_show(PREFIX + "class_graph.png")

# ==========================================================
# 7. CONFUSION MATRIX HEATMAP
# ==========================================================

plt.figure(figsize=(8, 6))
plt.imshow(cm)
plt.colorbar()
plt.title("Confusion Matrix Heatmap")
plt.xticks(range(len(label_encoder.classes_)), label_encoder.classes_, rotation=45)
plt.yticks(range(len(label_encoder.classes_)), label_encoder.classes_)

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xlabel("Predicted Class")
plt.ylabel("Actual Class")
save_and_show(PREFIX + "confusion_matrix_heatmap.png")

# ==========================================================
# 8. OPTIMIZATION GRAPH
# ==========================================================

plt.figure(figsize=(10, 6))
plt.plot(fitness_curve, marker="o")
plt.title("PSO Optimization Fitness Graph")
plt.xlabel("Iteration")
plt.ylabel("Best MSE")
plt.grid(True)
save_and_show(PREFIX + "optimization_graph.png")

# ==========================================================
# FINAL OUTPUT
# ==========================================================

print("\nAll results shown and saved successfully!")

print("\nGenerated CSV Files:")
print(classes_csv)
print(prediction_csv)
print(result_csv)

print("\nGenerated Graphs:")
print(PREFIX + "accuracy_graph.png")
print(PREFIX + "heatmap.png")
print(PREFIX + "comparison_graph.png")
print(PREFIX + "result_graph.png")
print(PREFIX + "prediction_graph.png")
print(PREFIX + "class_graph.png")
print(PREFIX + "confusion_matrix_heatmap.png")
print(PREFIX + "optimization_graph.png")

print("\nGenerated Model and Config Files:")
print(PREFIX + "regression_model.pkl")
print(PREFIX + "classification_model.pkl")
print(PREFIX + "scaler.pkl")
print(PREFIX + "label_encoder.pkl")
print(PREFIX + "summary.json")
print(PREFIX + "config.yaml")

print("\nOutput Folder:")
print(BASE_DIR)
```
